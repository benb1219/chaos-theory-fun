```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def find_steady_vals(r, x):
  data = []
  for i in range(500):
    x = r * x * (1 - x)
    data.append(x)
  vals = pd.Series(data).value_counts().head(100)
  return vals
```

```python
def find_steady(r, x):
  for i in range(500):
    x = r * x * (1 - x)
  return x
```

```python
r = 0.1
x_data = []
y_data = []
while(r < 4):
  results = find_steady_vals(r, .4)
  vals = results.index.tolist()
  for i in range(len(results)):
    y_data.append(vals[i])
    x_data.append(r)
  r += 0.001
plt.scatter(x_data, y_data, s=0.00001)
plt.xlabel("r")
plt.ylabel("steady-state value of x")
plt.figtext(0, -0.1, "Variation of the eventual population for the recursive equation X' = XR(1 - X), where X's initial state is 0.3")
plt.show()
```

```python
ax = plt.axes(projection='3d')

r = 1
x_data = []
y_data = []
z_data = []
while(r < 4):
  x = 0.3
  while(x < 0.6):
    results = find_steady_vals(r, x)
    vals = results.index.tolist()
    for i in range(len(results)):
      x_data.append(r)
      y_data.append(x)
      z_data.append(vals[i])
    x += 0.01
  r += 0.001
ax.scatter3D(x_data, y_data, z_data, s=0.0001)
```
